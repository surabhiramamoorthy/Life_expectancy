```python
from google.colab import files
uploaded=files.upload()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
data=pd.read_csv(io.BytesIO(uploaded['Life Expectancy Data.csv']))
print(data)
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
# Replacing the Null Values with mean values of the data
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)
data['Life expectancy ']=imputer.fit_transform(data[['Life expectancy ']])
data['Adult Mortality']=imputer.fit_transform(data[['Adult Mortality']])
data['Alcohol']=imputer.fit_transform(data[['Alcohol']])
data['Hepatitis B']=imputer.fit_transform(data[['Hepatitis B']])
data[' BMI ']=imputer.fit_transform(data[[' BMI ']])
data['Polio']=imputer.fit_transform(data[['Polio']])
data['Total expenditure']=imputer.fit_transform(data[['Total expenditure']])
data['Diphtheria ']=imputer.fit_transform(data[['Diphtheria ']])
data['GDP']=imputer.fit_transform(data[['GDP']])
data['Population']=imputer.fit_transform(data[['Population']])
data[' thinness  1-19 years']=imputer.fit_transform(data[[' thinness  1-19 years']])
data[' thinness 5-9 years']=imputer.fit_transform(data[[' thinness 5-9 years']])
data['Income composition of resources']=imputer.fit_transform(data[['Income composition of resources']])
data['Schooling']=imputer.fit_transform(data[['Schooling']])
```

```python

# Looking for null value in the data after replacing with mean value
data.isnull().sum()
```

```python
data=data.rename(columns = {'Country':'country',
                       'Year':'year',
                       'Status':'status',
                       'Life expectancy ':'lifeexpectancy',
                       'Adult Mortality':'mortality',
                       'infant deaths':'infantdeaths',
                       'Alcohol':'alchohol',
                       'percentage expenditure':'percentageexpediture',
                       'Hepatitis B':'hepatitis',
                       'Measles ':'measles',
                       ' BMI ':'bmi',
                       'under-five deaths ':'underfivedeaths',
                       'Polio':'polio',
                       'Total expenditure':'totalexpenditure',
                       'Diphtheria ':'diphtheria',
                       ' HIV/AIDS':'hiv/aids',
                       'GDP':'gdp',
                       'Population':'population',
                       ' thinness1-19years':'thinness_1_19_years',
                       ' thinness5-9 years':'thinness_5_9_years',
                       'Income composition of resources':'income_composition_of_resources',
                       'Schooling':'schooling'
                       })
print(data)

```

```python
#1.life expectancy vs year in India
#Isolating india
a=data[data.country=='India']
print(a)
plt.figure(figsize=(5,5))
plt.plot(a.lifeexpectancy,a.year)
plt.show
plt.xlabel('Life expectancy',fontsize=12)
plt.ylabel('year',fontsize=12)
plt.title('Life Expectancy vs year',fontsize=16)
```

```python
#2.life expectancy vs bmi in india
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(a.lifeexpectancy,a.bmi)

plt.xlabel("life expectancy")
plt.ylabel("bmi")
plt.title("life expectancy vs bmi")
plt.show()
```

```python
#3.scatterplot graph of schooling vs life expectancy for all countries
sns.scatterplot(data['lifeexpectancy'],data['schooling'])
```

```python
#4.countries having life expectancies above 80 and schooling above 20

data[data.schooling>20]
```

```python
#5.life expectancy vs population in india
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (5, 5))

# creating the bar plot
plt.bar(a.lifeexpectancy, a.population)

plt.xlabel("life expectancy")
plt.ylabel("population")
plt.title("life expectancy vs population")
plt.show()
```
